```python
#importing libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#loading the data
df = sns.load_dataset('iris')
```

```python
#dataset structure
#shape
print(df.shape)
#columns
print(df.columns)
#head
print(df.head())
```

```python
#summary of data
print(df.describe())
print(df.info())
```

```python
#visuallizations scatter plot
plt.figure(figsize=(6,4))
sns.scatterplot(data=df, x='sepal_length', y='petal_length', hue='species')
plt.title('Relationships Between Variables')
plt.show()
```

```python
#visualization Histrogram
plt.figure(figsize=(6,4))
sns.histplot(df['sepal_length'], bins=20, kde=True)
plt.title('Data Distribution')
plt.show()
```

```python
#visualization Box Plot(Outliers & Spread)
plt.figure(figsize=(6,4))
sns.boxplot(data=df, x='species', y='petal_length')
plt.title('Detect Outliers and Spread of Values')
plt.show()
```
